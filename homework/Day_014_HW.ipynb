```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '/Users/cindyhuang/Documents/Github/Data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train.groupby(["HOUSETYPE_MODE"]).AMT_CREDIT.mean()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train.HOUSETYPE_MODE.unique()

fig = plt.figure(figsize=(30,5))

for i in range(0, len(unique_house_type)):
    fig.add_subplot(1,len(unique_house_type),(i+1)).hist(app_train.AMT_CREDIT[app_train.HOUSETYPE_MODE == unique_house_type[i]])
    
    plt.title(str(unique_house_type[i]))
plt.show()
```
