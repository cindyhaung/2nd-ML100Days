```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.normal(scale=0.35, size=[10,10])
# matrix = np.random.random((10,10))*2-1

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.normal(scale=0.35, size=[nrow, ncol])
# matrix = np.random.random((nrow,ncol))*2-1

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = np.random.randn(nrow, ncol)
# matrix = np.random.randn(nrow,ncol)*2-1

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
